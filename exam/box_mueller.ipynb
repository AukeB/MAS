```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define box_muller method.
def box_muller(z1,z2):
    x1 = np.cos(2*np.pi*z2)*np.sqrt(-2*np.log(z1))
    x2 = np.sin(2*np.pi*z2)*np.sqrt(-2*np.log(z1))
    return x1,x2
```

```python
# Function that transforms Gaussian.
def change_gaussian(x1,x2):
    mu = 2.7; sigma = np.pi
    return x1*np.pi+mu,x2*np.pi+mu
```

```python
def gaussian(x,mu,sigma_squared):
    return (1)/(np.sqrt(2*np.pi*sigma_squared))*np.exp(-1*(x-mu)**2/(2*sigma_squared))
```

```python
# Generate 2 times 1000 random variables from a uniform distribution.
z1 = np.random.uniform(0,1,1000)
z2 = np.random.uniform(0,1,1000)
```

```python
# Transform the random arrays to Gaussian.
x1,x2 = box_muller(z1,z2)
```

```python
# Transform again.
y1,y2 = change_gaussian(x1,x2)
```

```python
# Add arrays to each other.
array = []
for i in range(len(y1)):
    array.append(y1[i]); array.append(y2[i])
```

```python
# Plot it in a normalized histogram.
plt.hist(array,density=True)
plt.show()
plt.close()
```

```python
# Compute analytical gaussian.
x = np.linspace(-7.5,12.5,1000)
y = gaussian(x,2.7,np.pi)
```

```python
# Plot the analytical curve
plt.plot(x,y)
plt.show()
plt.close()
```

```python
# Plot them both.
plt.hist(array,density=True)
plt.plot(x,y)
plt.show()
plt.close()
```
